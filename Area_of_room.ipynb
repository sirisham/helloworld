```python
#Area of a Room
'''
Write a program that asks the user to enter the width and length of a room. Once the values have been read, your program should compute and display the area of the room. The length and the width will be entered as floating point numbers. Include units in your prompt and output message; either feet or meters, depending on which
unit you are more comfortable working with.The length and the width will be entered as floating point numbers. Include units in your prompt and output message; either feet or meters, depending on which
unit you are more comfortable working with.
'''

print ("!!Compute Area of Room!!")
print ("Need 2 values \n1.Length\n2.Width")
length = float(input("Please enter the length(m):"))
width = float(input("Please enter the width(m):"))

area = length*width 

if length == width:
  print (f'Area of the room is {area}sqm') 
  print('Room is in square shape')
else :
  print (f'Area of the room is {area}sqm \nRoom is in rectangle shape')
```
