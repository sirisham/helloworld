```python
# Area of a Field
'''
Create a program that reads the length and width of a farmerâ€™s field from the user in
feet. Display the area of the field in acres.

Hint: There are 43,560 square feet in an acre.

'''
SQFT_PER_ACRE = 43560
#Details of the program
print ("!!Compute Area of Field in acres!!")
print ("Need 2 values \n1.Length\n2.Width")
#read user inputs
length = float(input("Please enter the length(feet):"))
width = float(input("Please enter the width(feet):"))
#compute the area of a feild
area =(length*width)/SQFT_PER_ACRE
#Display the output returing the area upto 3 decmal points
print (f'Area of field {area:.3f} acres')
```
