```python
# Libraries used 
import pandas as pd
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2

#Fuction to assign the expanded continent name from the code derived form the pycountry_convert library
def get_continent_name(con_name):
  if con_name == 'AS':
    name = 'ASIA'
  elif con_name == 'AF':
    name = 'AFRICA'
  elif con_name == 'EU':
    name = 'EUROPE'
  elif con_name == 'NA':
    name = 'NORTH-AMERICA'
  elif con_name == 'SA':
    name = 'SOUTH-AMERICA'
  elif con_name == 'AN':
    name = 'ANTARTICA'
  elif con_name == 'OC':
    name = 'OCEANIA'
  else :
    name = con_name
  return name
 #function to convert to alpah2 country codes and continents
def get_continent(country):
    try:
        cn_code =  country_name_to_country_alpha2(country)
    except:
        cn_code = 'Unknown' 
    try:
        cn_continent = country_alpha2_to_continent_code(cn_code)
    except:
        cn_continent = 'Unknown' 
    names = get_continent_name(cn_continent) 
    return (names)
 
# making data frame from excel

data = pd.read_excel('/content/sample_data/indicator hiv estimated prevalence% 15-49.xlsx') #Reading the input excel file into the pandas dataframe
continent_names = [] #empty list to hold the continent names

#print(type(data)) #verifying the type of the data variable 
  
# Converting a specific Dataframe
# column to list using list()
# function in Python
Name_List = list(data["Estimated HIV Prevalence% - (Ages 15-49)"])  

#Loop to find the continent name from the county information in the input file
for i in Name_List:
  con_name = get_continent(i)
  continent_names.append(con_name)

#Adding the continent imformation to the dataframe
data['continent'] = continent_names



#writes the dataframe to a csv file
data.to_csv('/content/sample_data/new_indicator hiv estimated prevalence% 15-49.csv',index = True)

data

```
