```python
import math
import numpy as np
import matplotlib.pyplot as plt

accuracy = []
pie = []
point=[ 10, 500, 1000, 5000, 10000, 50000, 80000, 100000, 200000]
#looping over point
for p in point:
    #Genereating the uniform float values 
    outX=np.random.uniform(0,1,p)
    outY=np.random.uniform(0,1,p)
    a= (outX-0.5)**2
    b= (outY-0.5)**2
    
    #Verifying if they fall inside the circle or outside the circle
    totalIn=a+b<0.25

    #Creating a inside points list from the above calculation
    inX=outX[totalIn]
    inY=outY[totalIn]
   
    #plotting
    fig=plt.figure(figsize=(10,10))
    axis = fig.add_subplot(1, 1, 1)
    circle=plt.Circle((0.5,0.5),radius=0.5,fill=False)
    rectangle=plt.Rectangle((0,0), 1, 1,edgecolor='k',fill=False)
    axis.add_patch(circle)
    axis.add_patch(rectangle)
    plt.title(f'Simulation Plot with {p} points')
    plt.scatter(outX,outY,color='blue',s=5)
    plt.scatter(inX,inY,color='pink',s=5)
    plt.show()
    
    # Calulating the Pie value here. using the sum method in numpy
    pi=4*np.sum(totalIn)/p
    
    #Appending it to the list for each iteration (Different P value)
    pie.append(pi)

    #absolute error value
    absolute_error = math.pi - pi

    #Appending the float absoulte value to plot the accuracy of the pie value in each iteration
    accuracy.append(math.fabs(absolute_error))

#Printing the pie value obtained in each iteration respectively
print('Pie values',pie)

#improvement per iteration
improvement_per_iteration = [accuracy[index] - accuracy[index-1] for index, value in enumerate(accuracy) if index > 0]

#improvement value line chart
print("\u0332".join("Improvements for iteration"))

plt.title('Improvement per iteration')
plt.plot(improvement_per_iteration)
plt.show()

```
