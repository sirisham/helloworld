```python
#UseCase : To identify the percentage of user who knows python in the each country responded to this survey

#Number of respondents knows pythion grouby country

country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').sum())
country_uses_python
```

```python


#number of respondents in each country

country_respondents = df_pb['Country'].value_counts()
country_respondents
```

```python
#Concating the above series to form a dataframe

python_df = pd.concat([country_respondents,country_uses_python],axis = 'columns')
python_df
```

```python
#Renaming the dataframe for better readability

python_df.rename(columns = {'Country' : 'NumberOfRespondents','LanguageWorkedWith' : 'NumKnowsPython'},inplace = True)
```

```python
#Adding a column to pythion dataframe that defines the percentage of people know python

python_df['PerKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumberOfRespondents'])*100
```
